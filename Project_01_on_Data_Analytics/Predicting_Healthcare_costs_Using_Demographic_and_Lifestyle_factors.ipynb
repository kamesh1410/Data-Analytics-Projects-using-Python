```python
#univariate analysis of object or categorical using 'pie chart'
category_counts=df['sex'].value_counts()
labels=category_counts.index
sizes=category_counts.values
colors=sns.color_palette('pastel')[0:len(labels)]
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.title('Gender Distribution')
plt.show()
```

```python
category_counts=df['region'].value_counts()
labels=category_counts.index
sizes=category_counts.values
colors=sns.color_palette('bright')[0:len(labels)]
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=90)
plt.title('Regions')
plt.show()
```

```python
category_counts=df['smoker'].value_counts()
labels=category_counts.index
sizes=category_counts.values
colors=sns.color_palette('bright6')[0:len(labels)]
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',startangle=90)
plt.title('Somkers or Not')
plt.show()
```
